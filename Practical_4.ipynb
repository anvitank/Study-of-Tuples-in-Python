```python
# Result Storing
'''A students exam result is stored in a tuple containing three fixed fields:
Subject name,marks obtained and grade
write a python program to :
1. store the exam result using a tuple
2. unpack the tuple elements into seperate variables.
3. Display the suject, marks, grade.
4. Check whether the student has scored 75 or more marks and print "Distinction" if the condition is satisfied.'''

student=("Maths",89,"A")
sub=student[0]
marks=student[1]
grade=student[2]
print("Subject Name:",sub)
print("Marks:",marks)
print("Grade:",grade)

if marks>75:
  print("Distinction")
```
